# Visualization (Pivot )

```python
pivot_table = data.pivot_table(index='hospital', columns='admission_type', values='billing_amount', aggfunc='sum')
pivot_table
```

```python
pivot_table = data.pivot_table(index='hospital', columns='admission_type', values='billing_amount', aggfunc='sum', fill_value=0)
pivot_table
```

```python
pivot_table = data.pivot_table(index=['hospital', 'gender'], columns='admission_type', values='billing_amount', aggfunc='sum')
pivot_table
```

```python
pivot_table = data.pivot_table(index='hospital', columns='admission_type', aggfunc='size')
pivot_table
```

```python
data['date_of_admission'] = pd.to_datetime(data['date_of_admission'])
pivot_table = data.pivot_table(index=pd.Grouper(key='date_of_admission', freq='M'), columns='admission_type', values='billing_amount', aggfunc='sum')
pivot_table                    
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the pivot table as a line chart
pivot_table.plot(kind='line')
plt.xlabel('Date of Admission')
plt.ylabel('Billing Amount')
plt.title('Billing Amount by Admission Type over Time')
plt.legend(title='Admission Type')
plt.xticks(rotation=45)
plt.show()
```
